# Choco Banana
---
image.png

If we shade a subset of the cells in the above grid, we partition the cells into orthogonally connected regions of shaded and unshaded cells. The goal of this puzzle is to shade the cells so that

*   all regions of shaded cells are rectangular,
*   all regions of unshaded cells are not rectangular, and
*   all number clues in the grid give the size of the region (shaded or unshaded) that the clue is in.

The answer to this puzzle is the product of the number of unshaded cells in each row.

Below the grid is a smaller example puzzle with solution and answer extraction.

# Solution (No Code)
---

Let yellow cells indicate unknown cells, blue indicated shaded and white indicate unshaded.

c1.jpg

Our strategy relies heavily on trial and error, with the following ideas used repeatedly throughout:

*   Cells with different numbers cannot be in the same connected part. In particular, if their edges come into contact, the colors must flip.
*   If a certain coloring of a cell connects adjacent components in an illegal way (such as forcing a blue component to be non rectangular, or to make a component too large, or join components that should be of different size), we can immediately color it the opposite color.
*   If a connected part has size equal to a number contained, we shade all adjacent cells the opposite color.
*   If we have a restrictive nearby surrounding, it is a good idea to go through possible shapes of a component quickly, especially if it is shaded. Often, only one of these shapes does not lead to a contradiction.

We start by immediately shading the 1s blue, as a size 1 component must be rectangular.

The 12 on the right side must be white, as there is no room for an area 12 rectangle.

We also see that the nearby 8 only one way to form an area 8 rectangle, allowing us to shade the following:

We can see that the 4s in the bottom right corner must be form a shaded square, and that the nearby 7 cannot be connected to the cell below the 1.

The 5 in the centre cannot be extended rightwards in a rectangle, as that would force the 4 above to be in a white block of size larger than 4, and it cannot be extended downwards due to the 8. We deduce that it is white, and hence the colors of the adjacent centre cells are also known.

Next, we can deduce that the centre 6 cannot be completely extended left, since the connected white cells to the top can have at most size 3, or completely extended downwards due to the 12, so it must have width and height at least 2. c3.jpg

The 3 in the centre cannot be extended left again, so we add a blue block there. That blue block cannot be connected to the block with the 6s, so we can deduce the shape of the block of 6s.

The 4 block cannot be extended completely upwards or rightwards, since the adjacent white components have size 3 and 5, so it must be 2x2.

The 11s clearly cannot be blue, so we shade them white.

On the group of 8s directly below it, we see that there is no room for an area 8 rectangular block due to it sharing a left edge with a size 5 component, hence we can shade all of them white.

From that, we see there is only one possible cell left to complete the size 5 component.

The 8 in the bottom-centre cannot be blue, so it must be white.

From there, we deduce that the adjacent 12 is shaded, and only one rectangle can take that area.

Shading the upper edge of the 12 block white gives us a white area of exactly 9 for the 9s above, allowing us to deduce that a shading for the nearby area from the adjacent cells needing to be shaded blue.

On the left, we notice there is no space for a shaded blue area of size 15, so we shade the 15 white and the 4 blue.

Trying to fit a 5x2 blue component with the rightmost 10 leads to a contradiction from the size of the top-centre 8 component, so it must be shaded white.
